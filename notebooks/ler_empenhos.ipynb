Visto que foram encontradas 1.317 duplicatas em (Ente, Ano, IdUnid, NrEmpenho), o código a seguir testa se acrescentar IdFonte resolve (indicando granularidade NE×Fonte):

```python
key_ne = ["Ente","Ano","IdUnid","NrEmpenho"]
key_ne_fonte = key_ne + ["IdFonte"]

tot = len(df)
uniq_ne       = df[key_ne].drop_duplicates().shape[0]
uniq_ne_fonte = df[key_ne_fonte].drop_duplicates().shape[0]

print("Únicos por NE.........:", uniq_ne, "de", tot)
print("Únicos por NE×Fonte...:", uniq_ne_fonte, "de", tot)

# Quantas duplicatas ainda sobram se incluirmos IdFonte?
dups_ne       = tot - uniq_ne
dups_ne_fonte = tot - uniq_ne_fonte
print("Duplicatas (NE).......:", dups_ne)
print("Duplicatas (NE×Fonte).:", dups_ne_fonte)
```

```python
# --- A) Quais colunas mais variam dentro das duplicatas de NE×Fonte? ---

base = ["Ente", "Ano", "IdUnid", "NrEmpenho", "IdFonte"]

# apenas as linhas que têm duplicatas para essa chave
dups = df[df.duplicated(subset=base, keep=False)].copy()

# (opcional) padronizar data; não é obrigatório para nunique()
# if "DtEmpenho" in dups.columns:
#     dups["DtEmpenho"] = pd.to_datetime(dups["DtEmpenho"], errors="coerce")

# candidatos a variar (filtra só os que existem no DF)
cand_all = ["Elemento","Cod_PT","NrProjAtiv","ProjAtiv","ProgTrab",
            "CPFCNPJCredor","CNPJRaiz","Tp_Empenho","IdContrato",
            "NrLicitacao","DtEmpenho"]
cand = [c for c in cand_all if c in dups.columns]

# fração de grupos em que cada coluna tem >1 valor (versão à prova de tipo)
vary_share = (dups.groupby(base)[cand]
                .agg(lambda s: s.nunique(dropna=False) > 1)  # True se varia no grupo
                .mean(axis=0)                                # fração de grupos
                .sort_values(ascending=False))

print("Fração de grupos (duplicados NE×Fonte) em que a coluna varia (top 10):")
print((vary_share * 100).round(1).head(10).astype(str) + "%")


# --- B) Classificação da “causa” das duplicatas: partição extra vs eventos ---

# colunas de valores/eventos (use apenas as que existem)
val_all = ["Vlr_SubEmpenho","Vlr_AnulacaoEmpenho","Vlr_Liquidacao","Vlr_Pagto",
           "Vlr_Empenho","Vlr_Anul_Liquidacao","Vlr_Empenhado","Vlr_Liquidado","Vlr_Pago"]
val_cols = [c for c in val_all if c in dups.columns]

def causa(gr):
    # meta = tudo que NÃO é chave base nem coluna de valores/eventos
    meta_cols = [c for c in gr.columns if c not in (base + val_cols)]
    # se QUALQUER metadado variar (>1 valor), então há partição extra
    meta_varia = gr[meta_cols].apply(lambda s: s.nunique(dropna=False) > 1).any()
    return "particao_extra" if meta_varia else "eventos_mesmo_NEfonte"

causas = dups.groupby(base).apply(causa).value_counts()
print("\nCausas das duplicatas (contagem de grupos NE×Fonte):")
print(causas)

# (opcional) veja alguns exemplos de grupos com partição extra
# ex_particao = dups.groupby(base).filter(
#     lambda g: g[[c for c in cand if c in g.columns]].nunique(dropna=False).gt(1).any()
# )
# ex_particao.sort_values(base).head(50)
```

```python
import itertools as it
import pandas as pd

# Base NE×Fonte
base = ["Ente","Ano","IdUnid","NrEmpenho","IdFonte"]

# Sugestão: normalize NrEmpenho como string com zeros (evita colisões futuras)
w = df["NrEmpenho"].astype(str).str.len().mode().iat[0]
df["NrEmpenho_norm"] = df["NrEmpenho"].astype(str).str.zfill(int(w))
base = ["Ente","Ano","IdUnid","NrEmpenho_norm","IdFonte"]

# Candidatas (filtra só as que existem)
cand_all = ["Cod_PT","ProgTrab","NrProjAtiv","ProjAtiv","Elemento","IdContrato","Tp_Empenho"]
cands = [c for c in cand_all if c in df.columns]

def n_unicos(cols): 
    return df.drop_duplicates(subset=cols).shape[0]

tot = len(df)
print("Duplicatas em NE×Fonte:", tot - n_unicos(base))

# 1) Ganancioso: adiciona 1 coluna por vez (a que mais reduz duplicatas)
key = base.copy()
cur = n_unicos(key)
added_steps = []

while cur < tot and cands:
    trials = [(c, n_unicos(key+[c])) for c in cands]
    c_best, n_best = max(trials, key=lambda x: x[1])
    if n_best == cur:
        break
    key.append(c_best)
    added_steps.append((c_best, tot - n_best))
    cands.remove(c_best)
    cur = n_best

print("\nChave sugerida (ganancioso):", key)
print("Duplicatas restantes:", tot - cur)
print("Passos (coluna → duplicatas restantes):", added_steps)

# 2) Se ainda restarem duplicatas, tente pares e trincas
left = tot - cur
if left > 0:
    pool = [c for c in cand_all if c in df.columns and c not in key]
    best_pairs = []
    for a,b in it.combinations(pool, 2):
        n2 = n_unicos(key + [a,b])
        best_pairs.append((a,b, tot - n2))
    best_pairs.sort(key=lambda t: t[2])
    print("\nTOP pares que mais reduzem duplicatas (col1, col2 → dup_restantes):")
    for a,b,r in best_pairs[:10]:
        print(f"{a:10s} {b:10s} → {r}")

    # (opcional) trincas — cuidado com custo
    best_triples = []
    for a,b,c in it.combinations(pool, 3):
        n3 = n_unicos(key + [a,b,c])
        best_triples.append((a,b,c, tot - n3))
    best_triples.sort(key=lambda t: t[3])
    print("\nTOP trincas (col1, col2, col3 → dup_restantes):")
    for a,b,c,r in best_triples[:10]:
        print(f"{a:10s} {b:10s} {c:10s} → {r}")
```
