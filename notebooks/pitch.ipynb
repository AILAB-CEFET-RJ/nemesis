## Gráfico de Fracionamento (várias notas abaixo do limite de licitação)

```python
import matplotlib.pyplot as plt

# Exemplo: 5 notas de R$ 7.900 no mesmo dia
valores = [7900, 7900, 7900, 7900, 7900]
dias = [f"Nota {i+1}" for i in range(len(valores))]

plt.figure(figsize=(8,4))
plt.bar(dias, valores, color="steelblue")
plt.axhline(8000, color="red", linestyle="--", label="Limite de licitação (R$8.000)")
plt.title("Exemplo de Fracionamento: Notas repetidas abaixo do limite", fontsize=12)
plt.ylabel("Valor (R$)")
plt.legend()
plt.tight_layout()
plt.savefig("fracionamento.png", dpi=300)
plt.show()
```

## Boxplot de Sobrepreço (preço do paracetamol)

```python
import matplotlib.pyplot as plt

# Valores simulados (em R$ por comprimido)
precos = [0.18, 0.20, 0.22, 0.19, 0.21, 0.20, 0.40]  # um município paga 0.40

plt.figure(figsize=(5,6))
plt.boxplot(precos, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.scatter(1, 0.40, color="red", s=120, zorder=3, label="Município fora da curva")
plt.title("Exemplo de Sobrepreço: Paracetamol 500mg", fontsize=12)
plt.ylabel("Preço por comprimido (R$)")
plt.legend()
plt.tight_layout()
plt.savefig("boxplot_paracetamol.png", dpi=300)
plt.show()
```

## Fracionamento (versão minimalista)

```python
import matplotlib.pyplot as plt

# Dados de exemplo: 5 notas de R$ 7.900 no mesmo dia
valores = [7900, 7900, 7900, 7900, 7900]
dias = [f"Nota {i+1}" for i in range(len(valores))]

plt.figure(figsize=(7,4))
bars = plt.bar(dias, valores, color="#1f77b4")

# Linha do limite de licitação
plt.axhline(8000, color="red", linewidth=3, linestyle="--")

# Rótulos
plt.title("Fracionamento de Notas", fontsize=18, weight="bold")
plt.ylabel("Valor (R$)", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Anotação visual do limite
plt.text(4.4, 8100, "Limite R$ 8.000", color="red", fontsize=12, weight="bold")

# Remover grid e bordas sup/dir
plt.grid(False)
plt.box(False)

plt.tight_layout()
plt.savefig("fracionamento_pitch.png", dpi=300)
plt.show()
```

## Sobrepreço (versão minimalista)

```python
import matplotlib.pyplot as plt

# Preços simulados (em R$ por comprimido)
precos = [0.18, 0.20, 0.22, 0.19, 0.21, 0.20, 0.40]  # outlier em 0.40

plt.figure(figsize=(4,6))
box = plt.boxplot(precos, vert=True, patch_artist=True,
                  boxprops=dict(facecolor="#1f77b4", linewidth=2),
                  medianprops=dict(color="black", linewidth=2),
                  whiskerprops=dict(linewidth=2),
                  capprops=dict(linewidth=2))

# Outlier destacado
plt.scatter(1, 0.40, color="red", s=200, marker="o", label="Município fora da curva")

# Rótulos
plt.title("Sobrepreço: Paracetamol 500mg", fontsize=16, weight="bold")
plt.ylabel("Preço por comprimido (R$)", fontsize=14)
plt.xticks([])
plt.yticks(fontsize=12)

# Remover grid e bordas sup/dir
plt.grid(False)
plt.box(False)

plt.legend(loc="upper right", fontsize=10)

plt.tight_layout()
plt.savefig("boxplot_paracetamol_pitch.png", dpi=300)
plt.show()
```
